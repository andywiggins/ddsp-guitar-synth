imports

```python
import numpy as np
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
from plot_listen.listen import play
from globals import *
from ddsp.harmonic_oscillator import HarmonicOscillator
from ddsp.sweetcocoa_harmonic import SweetCocoaHarmonicOscillator
```

# Create Harmonic Oscillator

```python
harm_osc = HarmonicOscillator()
sc_harm_osc = SweetCocoaHarmonicOscillator()
```

# Create Dummy Input Params

```python
z = {}

#### Test f0, harm_distr, and loudness

note_1 = torch.ones((32, 300)) * 440
note_2 = torch.ones((32, 300)) * 493.88
note_3 = torch.ones((32, 400)) * 554.37
z['f0'] = torch.concat((note_1, note_2, note_3), dim=-1)
z['harm_distr'] = torch.ones((32, 10, 1000)) * torch.tensor([1, 0, 0.8, 0, 0.6, 0, 0.5, 0.2, 0, 0.3])[None, :, None]
loud_1 = torch.ones((32, 500)) * 0.3
loud_2 = torch.ones((32, 500)) * 0.6
z['overall_amplitude'] = torch.concat((loud_1, loud_2), dim=-1)



# ## Test antialiasing

# note_1 = torch.ones((32, 500)) * 440
# note_2 = torch.ones((32, 500)) * 9000
# z['f0'] = torch.concat((note_1, note_2), dim=-1)
# z['harm_distr'] = torch.ones((32, 2, 1000)) * torch.tensor([1, 0.5])[None, :, None]
# z['loudness'] = torch.ones((32, 1000)) * 0.25
```

# Listen to harmonic oscillator

my implementation

```python
harm_audio = harm_osc(z)
play(harm_audio[0])
```

sweetcocoa implementation

```python
sc_harm_audio = sc_harm_osc(z)
play(sc_harm_audio[0])
```

# With inharmonicity added

```python
inharm_audio = harm_osc(z, string_inharm=0.012)
play(inharm_audio[0])
```
